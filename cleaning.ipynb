```python
df.dropna(inplace=True)
df.dropna(subset=['Fat','Fiber','Carbohydrates','Energy','Proteins','Salts','Sugars','Positive Points','Negative Points'], inplace=True)
```

```python
columns_to_convert = ['Fat', 'Fiber', 'Carbohydrates', 'Energy', 'Proteins', 'Salts', 'Sugars', 'Positive Points', 'Negative Points']
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')
df.dropna(subset=columns_to_convert, inplace=True)
```

```python
df.info()
```

```python

def is_suitable_for_blood_pressure(row):
    # Add your criteria here to determine suitability
    if row['Sugars'] <= 5 and row['Salts'] <= 1:
        return 1  # Suitable for blood pressure
    else:
        return 0  # Not suitable for blood pressure

def dia(row):
    if row['Sugars'] <= 6.5 :
        return 1 
    else:
        return 0  
    
def chol(row):
    if row['Fat'] < 14 :
        return 1 
    else:
        return 0  

df['Blood Pressure Suitability'] = df.apply(is_suitable_for_blood_pressure, axis=1)
df['Diabetes Suitability'] = df.apply(dia, axis=1)
df['Cholestrol Suitability'] = df.apply(chol, axis=1)


```

```python
df
```

```python

def skinny(row):
    if row['Fat'] <= 20 and row['Sugars'] <= 8 and row['Carbohydrates']<=80:
        return 1 
    else:
        return 0

def fit(row):
    if row['Fat'] <= 14 and row['Sugars'] <= 7 and row['Carbohydrates']<=55:
        return 1 
    else:
        return 0  
    
def obese(row):
    if row['Fat'] <= 10 and row['Sugars'] <= 5 and row['Carbohydrates']<=30:
        return 1 
    else:
        return 0  

df['Skinny'] = df.apply(skinny, axis=1)
df['Fit'] = df.apply(fit, axis=1)
df['Obese'] = df.apply(obese, axis=1)


```
