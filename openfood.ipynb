```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from openpyxl import Workbook, load_workbook
import time
import re
from selenium.common.exceptions import NoSuchElementException, WebDriverException

# Initialize Chrome WebDriver
driver = webdriver.Chrome()

# Load Excel workbook and sheet
wb = load_workbook("product_links.xlsx")
ws = wb.active

# Create a new Excel workbook and sheet for storing scraped data
output_wb = Workbook()
output_ws = output_wb.active
output_ws.append(["Product Name", "Fat", "Fiber", "Carbohydrates", "Energy", "Proteins", "Salts", "Calcium", "Cholesterol", "Sugars", "Additional Texts", "Positive Points", "Negative Points", "Nutritional Score"])

# Function to scrape product details
def scrape_product_details(url):
    driver.get(url)
    time.sleep(2)  # Add some delay to ensure all elements are loaded
    
    try:
        product_name = driver.find_element(By.CSS_SELECTOR, "h2.title-1").text
        
        nutrition_facts = {}
        nutrition_table_rows = driver.find_elements(By.CSS_SELECTOR, "#panel_nutrition_facts_table_content tbody tr")
        for row in nutrition_table_rows:
            key = row.find_element(By.CSS_SELECTOR, "td:nth-child(1)").text.strip()
            if key.lower() in ["fat", "fiber", "carbohydrates", "energy", "proteins", "salt", "calcium", "cholesterol", "sugars"]:
                value = row.find_element(By.CSS_SELECTOR, "td:nth-child(2)").text.strip()
                # Extract numerical values from the text using regular expression
                numeric_value = re.search(r'\d+(\.\d+)?', value)
                if numeric_value:
                    nutrition_facts[key] = numeric_value.group()
        
        # If any of the required nutritional facts are not found, set their value to None
        for fact in ["Fat", "Fiber", "Carbohydrates", "Energy", "Proteins", "Salt", "Calcium", "Cholesterol", "Sugars"]:
            if fact not in nutrition_facts:
                nutrition_facts[fact] = None
        
        # Scrape additional text
        additional_texts = []
        panel_group = driver.find_elements(By.CSS_SELECTOR, ".panel_group")
        for panel in panel_group:
            try:
                panel_text = panel.find_element(By.CSS_SELECTOR, ".panel_text").text.strip()
                additional_texts.append(panel_text)
            except NoSuchElementException:
                additional_texts.append("No additional text found")
        
        # Scrape positive points, negative points, and nutritional score using JavaScript to click
        nutrition_link = driver.find_element(By.XPATH, "//a[@href='#panel_nutriscore_content']")
        driver.execute_script("arguments[0].click();", nutrition_link)
        
        panel_nutriscore_content = WebDriverWait(driver, 10).until(EC.visibility_of_element_located((By.ID, "panel_nutriscore_content")))
        positive_points_element = panel_nutriscore_content.find_element(By.XPATH, "//p/strong[contains(text(), 'Positive points')]")
        negative_points_element = panel_nutriscore_content.find_element(By.XPATH, "//p/strong[contains(text(), 'Negative points')]")
        nutritional_score_element = panel_nutriscore_content.find_element(By.XPATH, "//p[contains(., 'Nutritional score')]")
    
        positive_points_text = positive_points_element.text
        negative_points_text = negative_points_element.text
        nutritional_score_text = nutritional_score_element.text
        
    except NoSuchElementException as e:
        print("An element was not found:", e)
        product_name = None
        nutrition_facts = {}
        additional_texts = "No additional text found"
        positive_points_text = None
        negative_points_text = None
        nutritional_score_text = None
    except WebDriverException as e:
        print("WebDriverException occurred:", e)
        return None, None, None, None, None, None
    
    return product_name, nutrition_facts, additional_texts, positive_points_text, negative_points_text, nutritional_score_text

# Loop through each row in the Excel sheet
for row in ws.iter_rows(min_row=201, max_row=440, values_only=True):
    product_name, nutrition_facts, additional_texts, positive_points, negative_points, nutritional_score = scrape_product_details(row[1])
    
    # If any of the values returned by scrape_product_details() are None, skip to the next iteration
    if product_name is None or nutrition_facts is None or additional_texts is None or positive_points is None or negative_points is None or nutritional_score is None:
        continue
    
    # Handle None values in nutrition_facts dictionary
    for key in nutrition_facts:
        if nutrition_facts[key] is None:
            nutrition_facts[key] = ""
    
    # Flatten the additional_texts list
    additional_text_string = "\n".join(additional_texts)
    
    # Write the scraped details to the output Excel sheet
    output_ws.append([product_name, nutrition_facts.get("Fat"), nutrition_facts.get("Fiber"), 
                      nutrition_facts.get("Carbohydrates"), nutrition_facts.get("Energy"), 
                      nutrition_facts.get("Proteins"), nutrition_facts.get("Salt"), 
                      nutrition_facts.get("Calcium"), nutrition_facts.get("Cholesterol"),
                      nutrition_facts.get("Sugars"), additional_text_string, positive_points, 
                      negative_points, nutritional_score])
    
# Save the output Excel file
output_wb.save("scraped_product_details.xlsx")

# Close the browser
driver.quit()
```
